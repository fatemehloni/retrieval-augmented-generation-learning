```python
from sentence_transformers import SentenceTransformer
import faiss
import numpy as np
```

```python
model = SentenceTransformer("all-MiniLM-L6-v2")
```

```python
texts = [
    "Artificial intelligence focuses on building systems that can perform tasks typically requiring human intelligence, such as reasoning, learning, and decision-making.",

    "Machine learning is a subfield of artificial intelligence that enables systems to learn patterns from data without being explicitly programmed.",

    "Deep learning is a subset of machine learning that uses multi-layer neural networks and has achieved strong results in image recognition and natural language processing.",

    "AI techniques are widely used in medical diagnosis, healthcare analytics, and clinical decision support systems to improve patient outcomes.",

    "In healthcare applications, artificial intelligence can assist doctors by analyzing medical images, predicting diseases, and personalizing treatment plans.",

    "Natural language processing allows machines to understand and generate human language, enabling applications such as chatbots, document search, and text summarization.",

    "Computer vision focuses on enabling machines to interpret visual information from images and videos, and is widely applied in autonomous vehicles and medical imaging.",

    "Football is a popular sport played worldwide, with millions of fans and professional leagues across different countries.",

    "Professional football requires physical fitness, tactical understanding, and teamwork, and is one of the most watched sports globally.",

    "Data quality and preprocessing play a critical role in the performance of machine learning and artificial intelligence systems."
]

len(texts)
```

```python
embeddings = model.encode(texts)
embeddings = np.array(embeddings).astype("float32")

embeddings.shape
```

```python
dimension = embeddings.shape[1]
index = faiss.IndexFlatL2(dimension)

index.add(embeddings)
index.ntotal
```

```python
# Define a query
query = "Applications of artificial intelligence in healthcare"

# Create embedding for the query
query_embedding = model.encode(query)
query_embedding = np.array([query_embedding]).astype("float32")

# Retrieve top-1 result
distances, indices = index.search(query_embedding, k=1)

distances, indices
```

```python
print(texts[indices[0][0]])
```

```python
# Retrieve top-3 results
distances, indices = index.search(query_embedding, k=3)

distances, indices
```

```python
for idx in indices[0]:
    print("-", texts[idx])
```

```python
# Retrieve top-5 results
distances, indices = index.search(query_embedding, k=5)

distances, indices
```
