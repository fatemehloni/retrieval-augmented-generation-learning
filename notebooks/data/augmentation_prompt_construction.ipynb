```python
# Retrieved chunks (simulated retrieval output)
retrieved_chunks = [
    "In healthcare applications, artificial intelligence can assist doctors by analyzing medical images, predicting diseases, and personalizing treatment plans.",
    "AI techniques are widely used in medical diagnosis, healthcare analytics, and clinical decision support systems to improve patient outcomes.",
    "Artificial intelligence focuses on building systems that can perform tasks typically requiring human intelligence, such as reasoning and learning."
]

# User question
question = "How is artificial intelligence used in healthcare?"
```

```python
# Prompt template for RAG-style augmentation
prompt = f"""
You are an assistant that answers questions using only the provided context.
If the answer is not contained in the context, say "I do not know."

Context:
{chr(10).join(f"- {chunk}" for chunk in retrieved_chunks)}

Question:
{question}

Answer:
"""
```

```python
print(prompt)
```

```python
# Split retrieved chunks by importance
core_chunks = retrieved_chunks[:1]        # most relevant
supporting_chunks = retrieved_chunks[1:]  # additional context

len(core_chunks), len(supporting_chunks)
```

```python
prompt_structured = f"""
You are an assistant that answers questions strictly based on the provided context.
If the answer cannot be inferred from the context, say "I do not know."

Core Evidence:
{chr(10).join(f"- {chunk}" for chunk in core_chunks)}

Supporting Context:
{chr(10).join(f"- {chunk}" for chunk in supporting_chunks)}

Question:
{question}

Answer:
"""
```

```python
print(prompt_structured)
```
